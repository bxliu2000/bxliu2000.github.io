```python
import torchvision.models as models
import torch
from PIL import Image
from torchvision import transforms
import matplotlib.pyplot as plt

size = 256
content_im = Image.open('./Berkeley.jpg')
style_im = Image.open('./abstract.jpg')
preprocess = transforms.Compose([
    transforms.Resize(size), 
    #transforms.CenterCrop(size),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
content_tensor = preprocess(content_im)
style_tensor = preprocess(style_im)

plt.imshow(content_tensor.permute(1, 2, 0))
plt.show()

plt.imshow(style_tensor.permute(1, 2, 0))
plt.show()
```

```python
def ContentLoss(x, y):
    """
    Takes in x and y as both tensors. 
    """
    return 0.5 * torch.sum(torch.square(x-y))

def gram(layer):
    F = layer.reshape(layer.shape[0], -1)
    return torch.matmul(F, F.T)

def StyleLoss(x, y):
    N_l2 = x.shape[0]**2
    M_l2 = (x.shape[1] * x.shape[2])**2
    return torch.sum(torch.square(gram(x)-gram(y))) / (4 * N_l2 * M_l2)
```

```python
inv_normalize = transforms.Normalize(
    mean=[-0.485/0.229, -0.456/0.224, -0.406/0.255],
    std=[1/0.229, 1/0.224, 1/0.255]
)
def visualize_step(canvas):
    canvas = canvas.detach().cpu().clone()
    canvas = inv_normalize(canvas)
    canvas = torch.clamp(canvas, 0, 1)
    canvas = canvas.permute(1, 2, 0).numpy()
    plt.imshow(canvas)
    plt.show()
    return canvas
```

```python
class FilterResponse:
    "Hook class for saving filter responses at layer x"
    
    def __init__(self):
        self.responses = []
        
    def __call__(self, module, module_in, module_out):
        self.responses.append(module_out)
    
    def clear(self):
        self.responses = []
        
    def __getitem__(self, idx):
        if type(idx) == tuple and len(idx) == 2:
            layer, i = idx
            return torch.squeeze(self.responses[layer])[i]
        elif type(idx) == int:
            return self.responses[idx]
        else:
            raise Exception('idx is either a tuple or a number.')
```

```python
ground = models.vgg16(pretrained=True).to('cuda')
ground_content = FilterResponse()
for layer in ground.modules():
    if isinstance(layer, torch.nn.modules.conv.Conv2d):
        layer.register_forward_hook(ground_content)
with torch.no_grad():
    output = ground(content_tensor.unsqueeze(0).to('cuda'))

ground = models.vgg16(pretrained=True).to('cuda')
ground_style = FilterResponse()
for layer in ground.modules():
    if isinstance(layer, torch.nn.modules.conv.Conv2d):
        layer.register_forward_hook(ground_style)
with torch.no_grad():
    output = ground(style_tensor.unsqueeze(0).to('cuda'))

assert len(ground_content.responses) > 0 and len(ground_style.responses) > 0
```

```python
from torch.optim import LBFGS

pred_content = FilterResponse()
pred_style = FilterResponse()
model_content = models.vgg16(pretrained=True).to('cuda')
model_style = models.vgg16(pretrained=True).to('cuda')
for layer in model_content.modules():
    if isinstance(layer, torch.nn.modules.conv.Conv2d):
        layer.register_forward_hook(pred_content)
for layer in model_style.modules():
    if isinstance(layer, torch.nn.modules.conv.Conv2d):
        layer.register_forward_hook(pred_style)
```

```python
N = 5
epochs = 20
lr = 1

a = 1e-3
b = 1e3
```

```python
canvas = torch.rand(content_tensor.shape, dtype=torch.float32, requires_grad=True, device='cuda')
theboos = [0, 2, 4, 7, 10]
coeff = [1e4, 0, 1e1, 0, 1e1, 0, 0, 1e1, 0, 0, 1e1]
optimizer = LBFGS([canvas], lr=lr)
def closure():
    #zero
    optimizer.zero_grad()
    pred_content.clear()
    pred_style.clear()
    #Forward pass
    model_content(canvas.unsqueeze(0))
    model_style(canvas.unsqueeze(0))

    #Content loss
    c_loss = ContentLoss(ground_content[N][0], pred_content[N][0])
    #Style Loss
    s_loss = 0
    for i in theboos[:N]:
    s_loss = s_loss + coeff[i] * StyleLoss(ground_style[i][0], pred_style[i][0])
    s_loss = s_loss / N
    #print(a * c_loss, b * s_loss)
    error = a*c_loss + b * s_loss
    error.backward()
    return error
for _ in range(epochs):
    optimizer.step(closure)
skio.imsave(str(N)+'-blend.jpg', visualize_step(canvas))
```
